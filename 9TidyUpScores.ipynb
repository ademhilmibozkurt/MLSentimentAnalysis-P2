## Tidy Up Scores

### Import Dependencies

```python
import pandas as pd
```

### Import Data

```python
df = pd.read_csv("8Scored.csv")
```

```python
df = pd.DataFrame(df)
```

### Change Rows

```python
for i in range(2,14):
    df.iloc[:, i] = df.iloc[:, i].apply(lambda x: "neg" if (x== "[0]" or x== "[[0]]") else "neu" if (x== "[1]" or x== "[[1]]") else "pos")
```
